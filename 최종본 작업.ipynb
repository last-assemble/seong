```python
rdf=me['revenue']>me['budget']
rdf=rdf.to_frame('calculation').astype(int)
me=me.merge(rdf, left_index=True, right_index=True, how='left')
me
```

# 학습 시키기

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers.core import Dense
from keras.layers import Dropout
import matplotlib.pyplot as plt
```

```python
x_data_genres = me.values[:, 24:39]
x_data_cast = me.values[:, 40:1040] # cast (캐스트)
x_data_spoken_langage = me.values[:,1041:1056] # spoken_langage (언어)
x_data_production_company = me.values[:,1060:] #production_company (제조 회사)
y_data = me.values[:, [-1]]


from sklearn.model_selection import train_test_split
genres_X_train, genres_X_test, genres_y_train, genres_y_test = train_test_split(x_data_genres, y_data, 
                                                    test_size=0.1, random_state=7)
cast_X_train, cast_X_test, cast_y_train, cast_y_test = train_test_split(x_data_cast, y_data, 
                                                    test_size=0.1, random_state=7)
spoken_langage_X_train, spoken_langage_X_test, spoken_langage_y_train, spoken_langage_y_test = train_test_split(x_data_spoken_langage, y_data, 
                                                    test_size=0.1, random_state=7)
production_company_X_train, production_company_X_test, production_company_y_train, production_company_y_test = train_test_split(x_data_production_company, y_data, 
                                                    test_size=0.1, random_state=7)
```

```python
np.random.seed(1)
print(genres_X_train.shape)
print(genres_X_test.shape)
print(genres_y_train.shape)
print(genres_y_test.shape)
print(cast_X_train.shape)
print(cast_X_test.shape)
print(cast_y_train.shape)
print(cast_y_test.shape)
print(spoken_langage_X_train.shape)
print(spoken_langage_X_test.shape)
print(spoken_langage_y_train.shape)
print(spoken_langage_y_test.shape)
print(production_company_X_train.shape)
print(production_company_X_test.shape)
print(production_company_y_train.shape)
print(production_company_y_test.shape)
print(genres_X_train.shape[1])
print(cast_X_train.shape[1])
print(spoken_langage_X_train.shape[1])
print(production_company_X_train.shape[1])
```

```python
def model_set(input_x):
    model = Sequential()
    model.add(Dense(256, input_shape=(input_x,), activation='relu'))

 #   model.add(Dropout(0.5))
  #  model.add(Dropout(0.5))
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='Adam', metrics=['accuracy'])
    model.summary()
    return model
```

# 장르 학습시키기

```python
model_genres = model_set(genres_X_train.shape[1])
hist_genres = model_genres.fit(genres_X_train, genres_y_train, validation_data=(genres_X_test, genres_y_test), epochs=100)

plt.figure(figsize=(12,8))
plt.plot(hist_genres.history['loss'])
plt.plot(hist_genres.history['val_loss'])
plt.plot(hist_genres.history['acc'])
plt.plot(hist_genres.history['val_acc'])
plt.legend(['loss','val_loss', 'acc','val_acc'])
plt.show()
```

# 캐스트 학습시키기

```python
model_cast = model_set(cast_X_train.shape[1])
hist_cast = model_cast.fit(cast_X_train, cast_y_train, validation_data=(cast_X_test, cast_y_test), epochs=100)

plt.figure(figsize=(12,8))
plt.plot(hist_cast.history['loss'])
plt.plot(hist_cast.history['val_loss'])
plt.plot(hist_cast.history['acc'])
plt.plot(hist_cast.history['val_acc'])
plt.legend(['loss','val_loss', 'acc','val_acc'])
plt.show()
```

# 언어 학습시키기

```python
model_spoken_langage = model_set(spoken_langage_X_train.shape[1])
hist_spoken_langage = model_spoken_langage.fit(spoken_langage_X_train, spoken_langage_y_train, validation_data=(spoken_langage_X_test, spoken_langage_y_test), epochs=100)

plt.figure(figsize=(12,8))
plt.plot(hist_spoken_langage.history['loss'])
plt.plot(hist_spoken_langage.history['val_loss'])
plt.plot(hist_spoken_langage.history['acc'])
plt.plot(hist_spoken_langage.history['val_acc'])
plt.legend(['loss','val_loss', 'acc','val_acc'])
plt.show()
```

# 제조 회사 학습시키기

```python
model_production_company = model_set(production_company_X_train.shape[1])
hist_production_company = model_production_company.fit(production_company_X_train, production_company_y_train, validation_data=(production_company_X_test, production_company_y_test), epochs=100)

plt.figure(figsize=(12,8))
plt.plot(hist_production_company.history['loss'])
plt.plot(hist_production_company.history['val_loss'])
plt.plot(hist_production_company.history['acc'])
plt.plot(hist_production_company.history['val_acc'])
plt.legend(['loss','val_loss', 'acc','val_acc'])
plt.show()
```
